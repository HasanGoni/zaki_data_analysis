```python
import pandas as pd
import numpy as np
from pathlib import Path
import os
from fastcore.all import *
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
data_path = Path(os.getenv('DATA_PATH'), 'zaki')
```

```python
def frm_sheet_name_to_df(
        sheet_name:str, 
        skiprows:int=7,
        year_filter:int=1981,
        first_name:str="Germany",
        last_name:str="Wartburgkreis (NUTS 2021)"):
    df = (
        pd.read_excel(
            data_path/'pat_ep_rtot_spreadsheet.xlsx', sheet_name=sheet_name,skiprows=skiprows, engine='openpyxl')

        .filter(regex='^(?!Unnamed).*([19|2]|TIME).*')
        )

    filtered_names = df.columns

    cols_ = [i for i in filtered_names if i == 'TIME' or (i.isdigit() and int(i) >= year_filter)]
    df = df[cols_]
    start_idx = df.query('TIME == @first_name').index[0]
    last_idx = df.query('TIME == @last_name').index[0]
    return df.loc[start_idx:last_idx,:]
```
